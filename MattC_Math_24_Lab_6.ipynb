```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-10,10,20)
```

```python
x
```

```python
m = 50
b = 17
```

```python
y = m * x + b
```

```python
plt.plot(x,y,'.') #plots given varibles
```

```python
r = 100*np.random.randn(20,)
```

```python
r
```

```python
plt.plot(r,'.') #Plots random values
```

```python
y2 = m * x + b + r
```

```python
plt.plot(x,y2,'.')
```

```python
m_guess = 1
b_guess = 1

y_guess = m_guess * x + b_guess

plt.plot(x,y_guess,'r.');
plt.plot(x,y2,'g.');
```

```python
for m_guess in np.linspace(-100,100,100):

        y_guess = m_guess * x

        plt.plot(x,y_guess,'r.');
        plt.plot(x,y2,'g.');
        plt.show()

        loss = np.sum((y_guess - y)**2)

        print(loss,'M = ',m_guess)

        #Tries out different slope values against actual plot values.
```

```python
best_loss = 100000000000
best_m = 0
best_b = 0

#Attempts to track best loss/slope

for m_guess in np.linspace(-100,100,100):
    for b_guess in np.linspace(-100,100,100):

        y_guess = m_guess * x + b_guess

        loss = np.sum((y_guess - y)**2)

        if loss < best_loss:

            best_loss = loss
            best_m = m_guess
            best_b = b_guess

#Loop which searches for slope and intercept plots with minimal loss.
```

```python
y_guess = best_m * x + best_b

plt.plot(x,y_guess,'r.');
plt.plot(x,y2,'g.');
plt.show()

print(best_loss,best_m,best_b)

#Showcases plot of best found values
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
def get_database(url,sheet_name):
    url_head = "https://docs.google.com/spreadsheets/d/"
    url_foot = "/gviz/tq?tqx=out:csv&sheet="
    url_body = url.split('/')[5]
    # sheet_name = "1"
    url_csv = url_head + url_body + url_foot + sheet_name
    df = pd.read_csv(url_csv, sep=',', skiprows=0)
    return df
```
